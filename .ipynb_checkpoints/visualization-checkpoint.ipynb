# CLASSIFICATION

##### Create rank_dict for renaming columns for future purposes in classification

```python
rankings = df['ranking'].value_counts().index.to_list()
# rankings.reverse()
# print(rankings)

rank_dict = {}
for i, el in enumerate(rankings):
    rank_dict[el] = i
rank_dict
```

```python
df['labels'] = df['ranking'].replace(rank_dict)
df
```

# Building Models

```python
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

#### 1. Splitting to X and y

```python
X = df.drop(['ranking', 'labels'], axis=1).to_numpy()
y = df['labels'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score
```

#### Splitting to TRAIN and TEST

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
models = [
#     {"model": LogisticRegression(random_state=1, n_jobs=-1),
#      "name": 'Logistic Regression',
#     "param_grid": {
#         "C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
#         "solver": ['liblinear', 'saga', 'newton-cg', 'lbfgs', 'sag'],
#         "penalty": ['l1', 'l2', 'elasticnet', 'none']
#         }
#     }
#     ,
#     {"model": Perceptron(random_state=1, n_jobs=-1),
#      "name": 'Perceptron',
#     "param_grid": {
#         "penalty": ['l2', 'l1', 'elastic'],
#         "alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10, 100],
#         'tol' : [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
#         }
#     }
#     ,
#     {"model": DecisionTreeClassifier(random_state=1),
#      "name": 'DTC',
#     "param_grid": {
#         "criterion": ['gini', 'entropy', 'log_loss'],
#         "splitter": ['best', 'random'],
#         "min_samples_split" : list(range(2, 10))
#         }
#     }
#     ,
    {"model": RandomForestClassifier(random_state=1, n_jobs=-1),
     "name": 'RFC',
    "param_grid": {
        'criterion': ['gini', 'entropy', 'log_loss'],
        'min_samples_leaf': list(range(1, 11)),
        'n_estimators': list(range(100, 1001, 100)),
        'max_depth': list(range(1, 21)), 
        'max_features': list(range(1, 21))
        }
    }
]
```

```python
base_result = []
for model in models:
    m = model['model']
    print(model['name'])
    %time m.fit(X_train, y_train)
    current_prediction = m.predict(X_test)
    score = accuracy_score(y_test, current_prediction)
    base_result.append({"name": model['name'], 
                       'estimator': m,
                       'score': score})
```

```python
base_result
```

```python
result = []
for model in models:
    rand_search = RandomizedSearchCV(model['model'], model['param_grid'], cv=5, n_iter=10, n_jobs=-1)
    %time rand_search.fit(X, y)
    result.append({"name": model['name'], 
                   'estimator': rand_search.best_estimator_,
                   'score': rand_search.best_score_})
```
